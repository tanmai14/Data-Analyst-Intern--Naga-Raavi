## Task 1: Clean the Salary Estimate Column and derive three new features from it min salary, max salary and mean salary

#### Before cleaning

```python
df['Salary Estimate'].head()
```

```python
temp=(df['Salary Estimate'][1].split(' ')[0].split('-'))
minSal=int(temp[0].replace('$','').replace('K','000'))
minSal
```

```python
df['min_salary']=np.zeros((len(df['Salary Estimate']),1))
df['max_salary']=np.zeros((len(df['Salary Estimate']),1))
df['avg_salary']=np.zeros((len(df['Salary Estimate']),1))
```

```python
for i in range(len(df['Salary Estimate'])):
    temp=(df['Salary Estimate'][1].split(' ')[0].split('-'))
    df['min_salary'][i]=int(temp[0].replace('$','').replace('K','000'))
    df['max_salary'][i]=int(temp[1].replace('$','').replace('K','000'))
    df['avg_salary'][i]=(df['min_salary'][i]+df['max_salary'][i])//2
```

```python
df.head()
```

#### After Cleaning the salary and generating it as new feature

```python
df.drop(labels=['Salary Estimate'],inplace=True,axis=1)
```

## Task 5: Cleaning the Revenue field

```python
df['Revenue'].value_counts().index
```

```python
df['Revenue'][1]
```

```python
df['Revenue'][1].split()[2].replace('$', '').replace('+', '')
```

```python
def min_revenue(revenue):
    if revenue == 'Unknown / Non-Applicable' or revenue == '-1':
        return -1
    elif revenue=='Less than $1 million (USD)':
        return 0
    elif revenue=='$10+ billion (USD)':
        return 10000000000
    elif 'million' in revenue and 'billion' in revenue:
        return int(revenue.split()[0].replace('$', '').replace('+', '')) * 1000000
    elif 'billion' in revenue:
        return int(revenue.split()[0].replace('$', '').replace('+', '')) * 1000000000
    elif 'million' in revenue:
        return int(revenue.split()[0].replace('$', '').replace('+', '')) * 1000000
    else:
        return 0
    
def max_revenue(revenue):
    if revenue == 'Unknown / Non-Applicable' or revenue == '-1':
        return -1
    elif revenue=='Less than $1 million (USD)':
        return 1000000
    elif revenue=='$10+ billion (USD)':
        return 10000000000
    elif 'million' in revenue and 'billion' in revenue:
        return int(revenue.split()[3].replace('$', '').replace('+', '')) * 1000000000
    elif 'billion' in revenue:
        return int(revenue.split()[2].replace('$', '').replace('+', '')) * 1000000000
    elif 'million' in revenue:
        return int(revenue.split()[2].replace('$', '').replace('+', '')) * 1000000
    else:
        return 0
df['min_revenue']=df['Revenue'].apply(min_revenue)
df['max_revenue']=df['Revenue'].apply(max_revenue)
```

```python
df.loc[:,['min_revenue','max_revenue']]
```

```python
df.drop(labels=['Revenue'],inplace=True,axis=1)
```

## Task 8: Identifying Job Title

```python
df['Job Title'].value_counts()
```

```python
def seniority(job_title):
    job_title=job_title.lower()
    snr=['sr','senior','lead','principal','vp','vice president','director','manager']
    for i in snr:
        if i in job_title:
            return 1
    if "jr" in job_title:
        return 0
    return -1
```

```python
df['Seniority']=df['Job Title'].apply(seniority)
```

```python
df.head()
```

### Grouping Job Titles

```python
def title_simplifier(title):
    title=title.lower()
    if 'data scientist' in title:
        return 'data scientist'
    elif 'data engineer' in title:
        return 'data engineer'
    elif 'stat' in title:
        return 'statistician'
    elif 'data analyst' in title:
        return 'data analyst'
    elif 'analyst' in title:
        return 'analyst'
    elif 'machine learning' in title:
        return 'machine learning'
    elif 'manager' in title:
        return 'manager'
    elif 'director' in title:
        return 'director'
    elif 'vice president' in title:
        return 'vp'
    else:
        return 'others'
df['Job_Group']=df['Job Title'].apply(title_simplifier)
```

```python
df.drop(labels=['Job Title'],axis=1,inplace=True)
```

```python
df.head()
```

## Task 10: One Hot encoding Skills from the Job Description

```python
df['python']=df['Job Description'].apply(lambda x: 1 if "python" in x.lower() else 0)
df['excel']=df['Job Description'].apply(lambda x: 1 if "excel" in x.lower() else 0)
df['sql']=df['Job Description'].apply(lambda x: 1 if "sql" in x.lower() else 0)
df['spark']=df['Job Description'].apply(lambda x: 1 if "spark" in x.lower() else 0)
df['cloud']=df['Job Description'].apply(lambda x: 1 if "cloud" in x.lower() else 0)
df['visualization']=df['Job Description'].apply(lambda x: 1 if ("visual" in x.lower() or "graph" in x.lower() or "dashboard" in x.lower()) else 0)
```

## Task 11: Cleaning the Size feature

```python
df['Size'].value_counts().index
```

```python
df['Size'][0].split()[0]
```

```python
def min_emp(emp):
    if emp == 'Unknown' or emp == '-1':
        return -1
    elif emp=='10000+ employees':
        return 10000
    return int(emp.split()[0])
    
def max_emp(emp):
    if emp == 'Unknown' or emp == '-1':
        return -1
    elif emp=='10000+ employees':
        return 10000
    return int(emp.split()[2])
df['min_emp']=df['Size'].apply(min_emp)
df['max_emp']=df['Size'].apply(max_emp)
```

```python
df.drop(labels=['Size'],axis=1,inplace=True)
```
